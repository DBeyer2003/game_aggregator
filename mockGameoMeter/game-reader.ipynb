```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


game_scores = pd.read_csv("/Users/DBeye/django_game/media/score_thingy.csv")

all_ids = []
all_vlists = []

for c in game_scores.columns:
  id_value = c
  values_list = game_scores[id_value].tolist()
  values_list = [x for x in values_list if str(x) != 'nan']
  all_ids.append(id_value)
  all_vlists.append(values_list)

print(all_ids)
print(all_vlists)
```

```python
score_list = pd.DataFrame({
    'id': all_ids,
    's_lists': all_vlists
})

score_list = score_list[~score_list['id'].str.contains('^Unnamed', na=False)]

score_list.to_csv('/Users/DBeye/django_game/media/meta_lists.csv',index=False)
```
